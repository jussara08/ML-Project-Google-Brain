```python
target = "HeartDisease"
print("\nTARGET DISTRIBUTION:\n", df[target].value_counts())
```

```python
num_cols = ["Age","RestingBP","Cholesterol","MaxHR","Oldpeak"]
import seaborn as sns
import matplotlib.pyplot as plt
for c in num_cols:
    x = df[c].dropna()

    # nice bins (age can be customized)
    if c == "Age":
        bins = np.arange(int(x.min()//5*5), int(x.max()//5*5)+6, 5)  # 5-year bins
    else:
        bins = 25

    plt.figure(figsize=(9,4.8))
    plt.hist(x, bins=bins)
    plt.title(f"Distribution of {c}")
    plt.xlabel(c)
    plt.ylabel("Count")
    sns.set_style("whitegrid")

    mean_ = x.mean()
    median_ = x.median()
    plt.axvline(mean_, linestyle="--", linewidth=2, label=f"Mean: {mean_:.2f}")
    plt.axvline(median_, linestyle=":", linewidth=2, label=f"Median: {median_:.2f}")
    plt.legend()

    plt.tight_layout()
    plt.show()
```

The age distribution shows a clear concentration in middle-aged and older individuals, with most observations clustered around the 5060 year range.  
Reflects a typical clinical population where cardiovascular risk increases with age 

MaxRH has fewer anomalities 

Resting BP distribution is moderately right-skewed. Most patients fall within normal to elevated ranges, while a smaller portion extends into higher values,which suggests that only a small number of patients have high blood pressure

```python
plt.figure(figsize=(9,4.8))
x_nz = df.loc[df["Cholesterol"] > 0, "Cholesterol"].dropna()
plt.hist(x_nz, bins=25)
plt.title("Cholesterol Distribution (excluding zeros)")
plt.xlabel("Cholesterol mg/dl")
plt.ylabel("Count")
sns.set_style("whitegrid")
plt.tight_layout()
plt.show()
```

Excluding the 0 values its possible to see a more typical distribution of cholesterol levels, with most patients having values in the 200300 mg/dl range. The distribution is right-skewed, indicating that while many patients have moderate cholesterol levels, there are some with significantly higher values. This suggests that while elevated cholesterol is common in the dataset, extremely high levels are less frequent but still present, which may be clinically relevant for cardiovascular risk assessment.

```python
cat_cols = ["Sex","ChestPainType","FastingBS","RestingECG","ExerciseAngina","ST_Slope"]

for c in cat_cols:
    tab = pd.crosstab(df[c], df[target]).sort_index()

    # A) stacked counts
    plt.figure(figsize=(10,5))
    x = np.arange(len(tab.index))
    y0 = tab[0].values if 0 in tab.columns else np.zeros(len(x))
    y1 = tab[1].values if 1 in tab.columns else np.zeros(len(x))

    plt.bar(x, y0, label="0 (No)")
    plt.bar(x, y1, bottom=y0, label="1 (Yes)")
    plt.xticks(x, tab.index.astype(str), rotation=30, ha="right")
    plt.title(f"{c} vs HeartDisease (Stacked Counts)")
    plt.xlabel(c)
    plt.ylabel("Count")
    sns.set_style("whitegrid")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # B) 100% stacked (row %)
    pct = tab.div(tab.sum(axis=1), axis=0) * 100
    y0p = pct[0].values if 0 in pct.columns else np.zeros(len(x))
    y1p = pct[1].values if 1 in pct.columns else np.zeros(len(x))

    plt.figure(figsize=(10,5))
    plt.bar(x, y0p, label="0 (No) %")
    plt.bar(x, y1p, bottom=y0p, label="1 (Yes) %")
    plt.xticks(x, pct.index.astype(str), rotation=30, ha="right")
    plt.title(f"{c} vs HeartDisease (100% Stacked)")
    plt.xlabel(c)
    plt.ylabel("Percent (%)")
    plt.ylim(0, 100)
    sns.set_style("whitegrid")
    plt.legend()
    # annotate top segment with %Yes
    for i, v in enumerate(y1p):
        plt.text(i, 50, f"Yes {v:.1f}%", ha="center", va="center", fontsize=9)
    plt.tight_layout()
    plt.show()
```

*Sex*: Heart disease appears more frequent among male patients.

ChestPainType: Certain chest pain categories show a higher proportion of heart disease.

ExerciseAngina: Patients with exercise-induced angina are more often in the heart disease group.

ST_Slope: Some ST slope patterns are strongly associated with heart disease.

RestingECG: Differences exist, but separation between groups is moderate.

FastingBS: Distribution looks similar across groups, showing weaker distinction.

***Conclusions***

**Age** 
Heart disease appears more common among older patients, with most affected individuals concentrated in middle to later adulthood. Age reflects a gradual increase in cardiovascular vulnerability rather than a sharp cutoff.

**RestingBP**
Patients show a wide range of resting blood pressure values, but higher readings appear more frequently among those with heart disease, suggesting increased cardiovascular strain in this group.

**Cholesterol (excluding 0)**
Cholesterol levels vary substantially between patients, with many heart disease cases occurring in moderate-to-high ranges. This indicates that elevated cholesterol is common among affected patients but not limited to extreme values.

**MaxHR**
Lower maximum heart rate levels tend to appear more often in patients with heart disease, which may reflect reduced exercise tolerance or underlying cardiac limitations.

**Oldpeak**
Most patients have low ST-depression values, but those with higher measurements are more frequently associated with heart disease, indicating that stronger stress responses of the heart are linked to worse outcomes.

**Sex**
Male patients represent a larger portion of the heart disease group, suggesting a noticeable gender difference in the studied population.

**ChestPainType**
Certain chest pain patterns align more strongly with heart disease, indicating that the nature of symptoms plays an important role in distinguishing patient risk profiles.

**ExerciseAngina**
Patients who experience angina during exercise show a higher presence of heart disease, highlighting exercise-related discomfort as a meaningful clinical signal.

**ST_Slope**
Specific ST slope categories are more dominant among heart disease patients, suggesting that changes in ECG patterns during stress are closely related to cardiac conditions.
 
 **RestingECG**
Resting ECG results show differences between groups, though the separation is less pronounced compared to exercise-related indicators.

**FastingBS**
Fasting blood sugar levels appear across both patient groups with less clear separation, implying a weaker standalone relationship with heart disease in this dataset.
 


```python
corr = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(9,7))
im = plt.imshow(corr.values, aspect="equal", cmap='Blues')
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right", color='black')
plt.yticks(range(len(corr.index)), corr.index)
plt.title("Correlation Heatmap (Numeric Features)")
plt.colorbar(im, fraction=0.046, pad=0.04)

#  correlation values (rounded)
for i in range(corr.shape[0]):
    for j in range(corr.shape[1]):
        plt.text(j, i, f"{corr.values[i,j]:.2f}", ha="center", va="center", fontsize=8)

plt.tight_layout()
plt.show()
```

****Patient-Focused Conclusions****

The analysis suggests that heart disease in this group appears more frequently among older and predominantly male patients, indicating that age and sex remain important characteristics of higher-risk profiles. Clinical indicators linked to physical stress on the heart  such as exercise-related symptoms and certain ECG or slope patterns  show noticeable differences between patients with and without heart disease, meaning these signs may reflect more advanced cardiovascular strain.

Cholesterol levels  and blood pressure vary  between patients, showing that risk is not concentrated in a single direction range but across diverse health conditions. Most patients have relatively low ST-depression values, while a smaller subset shows higher measurements, which may correspond to more severe cardiac stress.
